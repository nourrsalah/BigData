```python
df = df.dropna()
df = df.drop_duplicates()
df = df[df["state"] != "live"]
```

```python
df.isnull().sum()
```

```python
df["launched"] = pd.to_datetime(df["launched"])
df["deadline"] = pd.to_datetime(df["deadline"])
```

```python
df["CampaignDuration"] = (df["deadline"] - df["launched"]).dt.days
df[["launched","deadline","CampaignDuration"]].head()
```

```python
df["FundingRatio"] = df["usd_pledged_real"] / df["usd_goal_real"]
```

```python
def goal_category(goal):
    if goal < 10000:
        return "Low Goal"
    elif goal <= 100000:
        return "Medium Goal"
    else:
        return "High Goal"
df["GoalCategory"] = df["usd_goal_real"].apply(goal_category)
```

```python
df["LaunchMonth"] = df["launched"].dt.month_name()
```

```python
pipeline = [
    {
        "$group": {
            "_id": "$main_category",
            "total_projects": {"$sum": 1},
            "successful_projects": {
                "$sum": {
                    "$cond": [{"$eq": ["$state", "successful"]}, 1, 0]
                }
            }
        }
    },
    {
        "$project": {
            "success_rate": {
                "$multiply": [
                    {"$divide": ["$successful_projects", "$total_projects"]},
                    100
                ]
            }
        }
    }
]

list(collection.aggregate(pipeline))
```

```python
pipeline = [
    {
        "$group": {
            "_id": "$country",
            "avg_pledged": {"$avg": "$usd_pledged_real"}
        }
    },
    {"$sort": {"avg_pledged": -1}}
]

list(collection.aggregate(pipeline))
```

```python
pipeline = [
    {"$sort": {"FundingRatio": -1}},
    {"$limit": 5},
    {
        "$project": {
            "name": 1,
            "FundingRatio": 1,
            "usd_goal_real": 1,
            "usd_pledged_real": 1
        }
    }
]

list(collection.aggregate(pipeline))
```

```python
pipeline = [
    {"$match": {"state": "failed"}},
    {
        "$group": {
            "_id": "$category",
            "failures": {"$sum": 1}
        }
    },
    {"$sort": {"failures": -1}},
    {"$limit": 1}
]

list(collection.aggregate(pipeline))
```

```python
pipeline = [
    {
        "$group": {
            "_id": "$state",
            "avg_duration": {"$avg": "$CampaignDuration"}
        }
    }
]

list(collection.aggregate(pipeline))
```

```python
pipeline = [
    {"$match": {"state": "successful"}},
    {
        "$group": {
            "_id": "$LaunchMonth",
            "successful_projects": {"$sum": 1}
        }
    },
    {"$sort": {"successful_projects": -1}},
    {"$limit": 1}
]

list(collection.aggregate(pipeline))
```

```python
pipeline = [
    {
        "$group": {
            "_id": "$main_category",
            "avg_backers": {"$avg": "$backers"}
        }
    }
]

list(collection.aggregate(pipeline))
```

```python
pipeline = [
    {
        "$group": {
            "_id": None,
            "total_projects": {"$sum": 1},
            "reached_goal": {
                "$sum": {
                    "$cond": [{"$gte": ["$FundingRatio", 1]}, 1, 0]
                }
            }
        }
    },
    {
        "$project": {
            "percentage_reached_goal": {
                "$multiply": [
                    {"$divide": ["$reached_goal", "$total_projects"]},
                    100
                ]
            }
        }
    }
]

list(collection.aggregate(pipeline))
```

MongoDB Schema Design Justification

MongoDB follows a document-based model, so each Kickstarter project is stored as a single JSON document containing all attributes of the project. This design was chosen because the dataset naturally represents projects as independent records with multiple related fields such as category, goal, pledged amount, and campaign duration.

Storing the entire project in one document avoids joins and allows flexible querying across different attributes. Since MongoDB supports dynamic schemas, it was convenient to store both the original dataset attributes and the engineered features (CampaignDuration, FundingRatio, GoalCategory, and LaunchMonth) together in the same document. This structure makes aggregation queries such as success rates, average pledged amounts, and campaign statistics efficient using MongoDB aggregation pipelines.
